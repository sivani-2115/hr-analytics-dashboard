```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('data/raw/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Basic overview
print("Dataset Shape:", df.shape)

print("\nFirst 5 Rows:")
print(df.head())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nAttrition Count:")
print(df['Attrition'].value_counts())

print("\nAttrition Percentage:")
print(df['Attrition'].value_counts(normalize=True) * 100)
```

```python
# Check columns — understand what you're working with
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
dept_attr = df.groupby('Department')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
dept_attr.columns = ['Department', 'Attrition Rate (%)']

sns.barplot(data=dept_attr, x='Department', y='Attrition Rate (%)',
            palette='Reds_d')
plt.title('Attrition Rate by Department')
plt.tight_layout()
plt.savefig('../visuals/attrition_by_dept.png', dpi=150)
plt.show()
print(dept_attr)
```

```python
plt.figure(figsize=(10, 6))
role_attr = df.groupby('JobRole')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).sort_values(ascending=False).reset_index()
role_attr.columns = ['JobRole', 'Attrition Rate (%)']

sns.barplot(data=role_attr, x='Attrition Rate (%)', y='JobRole',
            palette='Oranges_d')
plt.title('Attrition Rate by Job Role')
plt.tight_layout()
plt.savefig('../visuals/attrition_by_role.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ot_attr = df.groupby('OverTime')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
ot_attr.columns = ['OverTime', 'Attrition Rate (%)']

sns.barplot(data=ot_attr, x='OverTime', y='Attrition Rate (%)',
            palette='coolwarm')
plt.title('Attrition Rate: Overtime vs No Overtime')
plt.tight_layout()
plt.savefig('../visuals/attrition_overtime.png', dpi=150)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',
            palette='Set2')
plt.title('Monthly Income: Left vs Stayed')
plt.tight_layout()
plt.savefig('../visuals/income_vs_attrition.png', dpi=150)
plt.show()

# Print averages
print(df.groupby('Attrition')['MonthlyIncome'].mean().round(0))
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',
            palette='Set2')
plt.title('Monthly Income: Left vs Stayed')
plt.tight_layout()
plt.savefig('../visuals/income_vs_attrition.png', dpi=150)
plt.show()

# Print averages
print(df.groupby('Attrition')['MonthlyIncome'].mean().round(0))
```

```python
plt.figure(figsize=(8, 5))
sat_attr = df.groupby('JobSatisfaction')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
sat_attr.columns = ['JobSatisfaction', 'Attrition Rate (%)']
sat_attr['JobSatisfaction'] = sat_attr['JobSatisfaction'].map(
    {1: '1-Low', 2: '2-Medium', 3: '3-High', 4: '4-Very High'}
)

sns.barplot(data=sat_attr, x='JobSatisfaction',
            y='Attrition Rate (%)', palette='YlOrRd_r')
plt.title('Attrition Rate by Job Satisfaction Level')
plt.tight_layout()
plt.savefig('../visuals/satisfaction_vs_attrition.png', dpi=150)
plt.show()
```

```python
df_pbi = df.copy()

# Convert Attrition to numeric for measures
df_pbi['Attrition_Num'] = df_pbi['Attrition'].map({'Yes': 1, 'No': 0})

# Age group bins — useful for Power BI slicers
df_pbi['Age Group'] = pd.cut(
    df_pbi['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60']
)

# Income band
df_pbi['Income Band'] = pd.cut(
    df_pbi['MonthlyIncome'],
    bins=[0, 3000, 6000, 10000, 20000],
    labels=['Low (<3K)', 'Mid (3-6K)', 'High (6-10K)', 'Very High (10K+)']
)

# Satisfaction labels
sat_map = {1: '1 - Low', 2: '2 - Medium', 3: '3 - High', 4: '4 - Very High'}
df_pbi['JobSatisfaction Label'] = df_pbi['JobSatisfaction'].map(sat_map)
df_pbi['WorkLifeBalance Label'] = df_pbi['WorkLifeBalance'].map(sat_map)

# Save
df_pbi.to_csv('data/processed/hr_cleaned.csv', index=False)
print("Done. Shape:", df_pbi.shape)
```
